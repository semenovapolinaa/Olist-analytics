# Финальные инсайты и рекомендации

## 1. Ключевые инсайты:

- Опоздания доставки ухудшают пользовательский опыт

Средние оценки заказов ниже, доля негативных отзывов выше у опоздавших заказов.

- Опоздавшие заказы имеют более высокую медианную стоимость

Это связано с более сложной логистикой (объем, вес, расстояние), что повышает риск задержки.

- Влияние опозданий на деньги и пользовательскийц опыт одновременно

Даже дорогие заказы, которые важны для дохода, при опоздании получают низкие оценки — бизнес теряет на обоих фронтах.

- Массовые сбои видны на уровне месяца

Пример: октябрь 2016 г. — доля опозданий почти 100%, что явно указывает на массовые логистические проблемы.

- Сильная статистическая разница между группами

A/B анализ подтверждает, что различия в оценках, доле плохих отзывов и медианной стоимости заказов значимы (p-value < 0.05).

- Доля опозданий и оценки хорошо коррелируют

Данные визуально и статистически показывают, что рост доли опоздавших заказов сопровождается снижением пользовательского опыта.

## 2. Бизнесовые рекомендации

- Сократить количество опозданий

- Приоритизация дорогих заказов

Так как задержки влияют на наиболее ценные заказы, логистика должна уделять особое внимание таким заказам.

- Работа с клиентским опытом

Коммуникация при задержках, компенсации, уведомления о статусе доставки, чтобы смягчить негативные отзывы.

- Мониторинг и аналитика в реальном времени

Использовать KPI по доле опоздавших заказов, median AOV, и bad review rate для контроля работы логистики.

## 3. Ограничения анализа

- Данные исторические (2016–2018)

Могут отличаться от текущей ситуации, поэтому прямое применение выводов требует актуализации.

- Отсутствие внешних факторов

Сезонность, праздники, региональные сбои, маркетинговые кампании — не учитывались.

- Агрегации и округления

Для визуализаций оценки были округлены, но статистический анализ проводился на исходных данных.

- Данные симулированы

Некоторые выводы основаны на смоделированных паттернах; в реальной практике могут быть отклонения.
