# Olist Feature Engineering

## 1. Цель этапа

Cоздать аналитическую таблицу на уровне заказа, которую можно:
- Использовать в Python
- Визуализировать в Superset
- Применять для A/B теста

## 2. Принцип формирования фичей

Формирование фичей основывалось на бизнес-контексте маркетплейса. Для анализа были выделены ключевые направления: доставка, клиентский опыт, финансовые показатели и структура заказа. Выполнена агрегация до уровня 1 строка = 1 заказ.

Перед тем, как формировать аналитическую таблицу, выделю фичи, на основе которых будет формироваться таблица и которые я считаю актуальными для данной задачи:

**Клиентский опыт:**
- reviews_count
- avg_review_score
- has_review

**Доставка:**
- delivery_days
- is_late_delivery
- total_freight

**Финансовые показатели:**
- items_total_price
- total_payment
- max_installments
- total_amount_olist

**Структура заказа:**
- items_cnt
